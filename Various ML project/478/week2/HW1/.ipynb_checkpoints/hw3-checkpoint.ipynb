### b, standard linear regression

```python
#standard linear regression
def standRegres(xArr,yArr):
    xMat = np.mat(xArr); yMat = np.mat(yArr).T
    xTx = xMat.T*xMat
    if np.linalg.det(xTx) == 0.0:
        print("This matrix is singular, cannot do inverse")
        return
    ws = xTx.I * (xMat.T*yMat)
    return ws
#mat is used to interpret the input as matrix
```

```python
ws=standRegres(x,y)
ws
```

```python
#RMSE
def rmse(xArr,yArr):
    ws=standRegres(xArr,yArr)
    xMat = np.mat(xArr); yMat = np.mat(yArr).T
    yHat=xMat*ws
    a=0
    for act,prd in zip(yMat,yHat):
        a=a+(act-prd)**2
    rmse=np.sqrt(a/len(yHat))
    return rmse    
```

```python
rmse(x,y)
```

```python
#RMSE_cv
from sklearn.model_selection import KFold
def rmse_cv(xArr,yArr,cv):
    kf = KFold(n_splits=cv)
    rmse=0
    for train_index, test_index in kf.split(xArr, yArr):
        X_train, X_test = xArr[train_index], xArr[test_index]
        y_train, y_test = yArr[train_index], yArr[test_index]
        beta=standRegres(X_train,y_train)
        p=beta.T*X_test.T
        p=p.T
        a=0
        for act,prd in zip(y_test,p):
            a=a+(act-prd)**2
        rmse+=np.sqrt(a/len(yHat))
    return rmse/cv   
```

## 2, Automatic Document Clustering

### a, distance functions

```python
from sklearn.model_selection import train_test_split
import scipy
```

```python
def Euclidean(x,y):
    #x=x.T
    x=np.array(x)
    #y=y.T
    y=np.array(y)
    dists = np.sqrt(((x - y)**2).sum(axis=0))
    return dists
```

```python
def Cosine(x,y):
    x_norm = np.array([np.linalg.norm(x[i]) for i in range(len(x))])
    y_norm = np.linalg.norm(y)
    sims=np.dot(x,y)/(x_norm*y_norm)
    dists=1-sims
    return dists
```

### c, Kmeans clustering function

```python
def randCent(dataSet,k):
    n = dataSet.shape[1] #n is the number of attribute (terms)
    centroids = np.zeros((k,n), dtype=float) # create the k centroids
    for j in range(n): #create random cluster centers
        minJ = min(dataSet[:,j])
        rangeJ = float(max(dataSet[:,j]) - minJ)
        centroids[:,j] = minJ + rangeJ * np.random.rand(k)
    return centroids 
```

```python
def kMeans(dataSet, k, distMeas=Euclidean, createCent=randCent):
    m = dataSet.shape[0]
    clusterAssment = np.zeros((m,2))#create mat to assign data points 
                                      #to a centroid, also holds SE of each point
    centroids = createCent(dataSet, k)
    clusterChanged = True
    while clusterChanged:
        clusterChanged = False
        for i in range(m):#for each data point assign it to the closest centroid
            minDist = np.inf; minIndex = -1
            for j in range(k):
                if distMeas == Euclidean:
                    distJI = Euclidean(centroids[j,:],dataSet[i,:])
                elif distMeas == Cosine:
                    distJI = scipy.spatial.distance.cosine(centroids[j,:],dataSet[i,:])
                if distJI < minDist:
                    minDist = distJI; minIndex = j
            if clusterAssment[i,0] != minIndex: clusterChanged = True
            clusterAssment[i,:] = minIndex,minDist**2
        for cent in range(k):#recalculate centroids
            ptsInClust = dataSet[np.nonzero(clusterAssment[:,0]==cent)[0]] #get all the point in this cluster - Note: this was incorrect in the original distribution.
            if(len(ptsInClust)!=0):
                centroids[cent,:] = np.mean(ptsInClust, axis=0) #assign centroid to mean - Note condition was added 10/28/2013
    return centroids, clusterAssment
```
